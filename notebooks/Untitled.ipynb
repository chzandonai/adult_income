```python
import pandas as pd
import numpy as np
```

```python
from dateutil.relativedelta import relativedelta
```

```python
def add_mob(anomes, meses=1):
    data = pd.to_datetime(anomes, format='%Y%m') + relativedelta(months=meses)
    return data.year*100 + data.month

def get_mes_list(inicio, fim):
    data = inicio
    lista = [data]
    while data < fim:
        data = add_mob(data)
        lista.append(data)
    return lista
```

```python
df = pd.read_csv("../data/adult.csv")
```

```python
rs = np.random.RandomState(32)

df['anomes'] = rs.choice(get_mes_list(202107,202306), df.shape[0])
```

```python
df = df.sort_values(by='anomes')
```

```python
for anomes in df.anomes.unique():
    print(anomes)
    aux = df.loc[df.anomes==anomes]
    aux.to_csv(f'../data/adult_{anomes}')
```
